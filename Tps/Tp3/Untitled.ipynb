```python
import numpy as np

# Definir una matriz cuadrada
A = np.array([[2, 1],
              [1, 2]])

# Calcular valores propios y vectores propios
valores_propios, vectores_propios = np.linalg.eig(A)

print("Valores propios:", valores_propios)
print("Vectores propios:\n", vectores_propios)
n = len(valores_propios)
determinante = 1
for i in range(n):
    determinante *= valores_propios[i]
print(determinante)
```

```python
import Lineales as li
import numpy as np

A = [[2,-1,-1],
     [0,2,1],
     [-1,1,0.5]]

b = [0,3,0.5]

def Eliminacion_Gauss(A,
                      b,
                      ):
    """
    A: Matriz
    B: Vector
    """
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    n = len(b)
    M = np.concatenate((A,b.reshape(n,1)), axis=1) # matriz ampliada (A|b)
    print(M)
    print("")
    for k in range(n-1):
        if M[k,k] == 0:
            raise ValueError(f"Pivote nulo en la fila: {k}. el metodo falla")
        for i in range(k+1, n):
            factor = M[i,k] / M[k,k]
            for j in range(k,n+1):
                M[i,j] -= factor*M[k,j]     # Construyendo matriz U
            M[i,k] = factor #    construyendo matriz L
    print(M)

    x = np.zeros(n)
    for i in range(n-1,-1,-1):
        suma_conocida = np.dot(M[i,i+1:n], x[i+1:n])
        x[i] = (M[i,n] - suma_conocida) / M[i,i]
    return x

x = Eliminacion_Gauss(A,b)
print(f"vector: {x}")
```
