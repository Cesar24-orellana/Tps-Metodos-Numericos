```python
import matplotlib.pyplot as plt
import numpy as np

puntos = [0.5,0.625,0.125,0.25,1.01,0.15625,0.75,1.5,0.1875,0.375,0.875,1.75,0.21875,0.4375,0.3125,0.21875]

# arreglo de ceros para el tamaño
y = np.zeros(len(puntos))

# Tamaño del gráfico
plt.figure(figsize=(15,2))

# Gráfico
plt.plot(puntos, y, 'go')

# 2. Ocultamos los bordes (spines) de arriba, derecha e izquierda
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

# 3. Nos aseguramos que la línea de abajo pase exactamente por y=0
plt.gca().spines['bottom'].set_position('zero')

# Opcional: Configuramos las marquitas del eje X (ticks) para ver bien la escala
plt.xticks(np.arange(0, max(puntos) + 0.5, 0.125))

# Le ponemos un título
plt.title("Representación de números en punto flotante (b=2, t=3, e=2)")

# Mostramos el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f_original(x):
    return x - np.sin(x)

def f_nueva(x):
    return x**3 / 6 - x**5 / 120

eps = 1e-3
x = np.linspace(0 - eps, 0 + eps, 100, dtype = np.float32)

plt.plot(x, f_original(x), label="original")
plt.plot(x, f_nueva(x), label="nueva")
plt.legend()
```

```python
# 6
import numpy as np

def serie_og(n):
    resultado = 0
    for i in range(1, n + 1):
        resultado = resultado + (1 / i**4)
    return resultado
    
def serie_ascendente(n):
    resultado = np.float32(0.0)
    for i in range(1, n + 1):
        resultado = np.float32(resultado + (1 / i**4))
    return resultado

def serie_descendente(n):
    resultado = np.float32(0.0)
    for i in range(n, 0, -1):
        resultado = np.float32(resultado + (1 / i**4))
    return resultado

n = 10000
valor_real = np.pi**4 / 90 
og = serie_og(n)
asc = serie_ascendente(n)
desc = serie_descendente(n)

er_a = np.abs(valor_real - asc) / valor_real
er_d = np.abs(valor_real - desc) / valor_real

print(f"Error relativo calculado de forma ascendente: {er_a}")
print(f"Error relativo calculado de forma descendente: {er_d}")
print("Esto se debe a que en el caso descendente sumo términos de magnitud similar, minimizando la pérdida de dígitos significativos.")
        
```

```python
#7
import numpy as np
import math as m

def aprox_uno(x, n):
    resultado = 0
    for i in range(n):
        resultado = resultado + (-x)**i / m.factorial(i)
        return resultado

def aprox_dos(x, n):
    resultado = 0
    for i in range(n):
        resultado += x**i / m.factorial(i)
    resultado = 1 / resultado
    return resultado

x_dado = 8.3
valor_real = 2.485168e-4

print(f"Valor Verdadero: {valor_real}")
print(f"Serie 1 (4 térm): {aprox_uno(x_dado, 4)}")
print(f"Serie 2 (4 térm): {aprox_dos(x_dado, 4)}")

# pruebo termino grande
print(f"Serie 1 (20 térm): {aprox_uno(x_dado, 20)}")
print(f"Serie 2 (20 térm): {aprox_dos(x_dado, 20)}")

print(f"Serie 1 (2 térm): {aprox_uno(x_dado, 2)}")
print(f"Serie 2 (2 térm): {aprox_dos(x_dado, 2)}")

print("La primera aproximación acumula error por tener restas, y 'cancela' las cifras significativas")
```

```python
#8
import numpy as np

def nueva(x):
    return (np.e**x - np.e**(-x)) / 2 * x

# imprimo con distintos valors
for i in range(1, 21, 2):
    x_val = 10**-i 
    resultado = nueva(x_val)
    print(f"Para x = 1e-{i:02d} -> Resultado: {resultado:.18f}")

# cuando x es muy chico, el numerador tiende a 0, pero se pierden cifras significativas pues cada vez es más cercano a uno.
```
