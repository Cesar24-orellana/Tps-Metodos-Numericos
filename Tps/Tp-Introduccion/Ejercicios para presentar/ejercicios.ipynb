## Ejercicio: Sucesión de Fibonacci

Implementa una función que devuelva una lista con los primeros `n` términos de la sucesión de Fibonacci.

**Especificaciones:**
- Define una función llamada `fibonacci(n)` que reciba un número entero no negativo como parámetro y devuelva una lista con n elementos.
- La función debe validar que n sea mayor o igual a 0.
- Los primeros dos números de la secuencia son 0 y 1.
- Los siguientes términos de la secuencia son el resultado de sumar los dos anteriores.
- Utiliza **Python puro** sin librerías externas
- Optimiza la solución (preferentemente evitar recursión sin memoización para valores grandes)

**Ejemplo:** `fibonacci(6)` deberá retornar `8` (sucesión: 0, 1, 1, 2, 3, 5, 8)

````python
# Hacer
def fibonacci(n):
    Lista = [0, 1]
    if n>=0:
        for i in range(n):
            Lista.append(Lista[i+1]+Lista[i])
    return Lista

nuevaLista = fibonacci(6)
print(nuevaLista)
````

## Ejercicio: Factorial

Implementa una función que calcule el factorial de un número entero.

**Especificaciones:**
- Define una función llamada `factorial(n)` que reciba un número entero positivo como parámetro
- La función debe validar que el número sea mayor o igual a 0
- Retorna el factorial de n (n! = n × (n-1) × (n-2) × ... × 1)
- Utiliza **Python puro** sin librerías externas
- Intenten hacer dos versiones: con bucles y con recursión

**Ejemplo:** `factorial(5)` deberá retornar `120` (5*4*3*2*1)

````python
# Hacer
def factorial(n):
    if n >= 0:
        if n > 0:
            return n*factorial(n-1)
        else :
            return 1
    return n

def factorial2(n):
    if n >= 0:
        num = 1
        while n > 1:
            num *= n
            n = n-1
        return num
    return n
    
print(f"Resultado de Factorial en bucle de 5 = {factorial2(5)} | Resultado de Factorial en recursiva de 5 = {factorial(5)}")
````

## Ejercicio: Producto Escalar de Dos Vectores

Implementa una función que calcule el producto escalar (dot product) de dos vectores.

**Especificaciones:**
- Define una función llamada `producto_escalar(vector1, vector2)` que reciba dos listas como parámetros
- La función debe validar que ambos vectores tengan la misma longitud
- Retorna el resultado del producto escalar (suma de los productos de los elementos correspondientes)
- Utiliza **Python puro** (sin librerías externas como NumPy)
- Implementa la lógica usando únicamente funciones y estructuras básicas de Python

**Ejemplo:** `producto_escalar([1, 2, 3], [4, 5, 6])` deberá retornar `32` `(1*4 + 2*5 + 3*6)`

````python
# Hacer
def producto_escalar(vector1, vector2):
    if len(vector1) == len(vector2):
        producto = 0
        i = 0
        while i < len(vector1):
            producto += (vector1[i] * vector2[i])
            i = i+1
        return producto
    return 0

print(producto_escalar([1, 2, 3], [4, 5, 6]))
````

## Ejercicio: Norma de un Vector

Implementa una función que calcule la norma (magnitud) de un vector.

**Especificaciones:**
- Define una función llamada `norma(vector)` que reciba una lista como parámetro
- La norma euclidiana (vista en algebra lineal) se calcula como: ||v|| = √(v₁² + v₂² + ... + vₙ²)
- Retorna el valor de la norma
- Utiliza **Python puro** sin librerías externas
- Tu función principal `norma()` debe usar la norma euclidiana por defecto

**Ejemplo:** `norma([3, 4])` deberá retornar `5.0`

````python
# Hacer
def norma(vector):
    suma = 0
    for i in range(len(vector)):
        suma += vector[i]**2
    return suma**0.5

print(norma([3, 4]))
````

## Ejercicio: Divisores de un número

Implemente una función que reciva un número y devuelva una lista con sus divisores.

**Especificaciones**:

* Los divisores de un número son aquellos números enteros que lo dividen de forma exacta, es decir, sin dejar residuo o resto cero.
* Defina una función llamada `encontrar_divisores(num)` que reciba un número y devuelva una tupla o lista.
* Los divisores siempre inlcuyen al 1 y al número en cuestión.

**Ejemplo**: `encontrar_divisores(10)` deberá retornar `(1, 2, 5, 10)`

````python
# Hacer
def encontrar_divisores(num):
    if num == 0 : return num  # Experimentando con la tipografia(0 escritura o como se diga) de python
    divisores = []
    for i in range(num):
        if (num % (num-i)) == 0: 
            divisores.append(num//(num-i))
    return divisores

nuevaList = encontrar_divisores(10)
print(nuevaList)
````

## Ejercicio: Es primo

Implemente una función que reciva un número y devuelva un booleano indicando si es primo.

**Especificaciones**:

* Los números primos son aquellos que solo se pueden dividir entre 1 y sí mismos.
* Defina una función llamada `es_primo(num)` que reciba un número y devuelva una booleano.

**Ejemplo**: `es_primo(5)` deberá retornar `True`

````python
# Hacer
def es_primo(num):
    if num <= 0 : return False
    contador = 0
    for i in range(num):
        if (num%(i+1))==0: 
            contador += 1
    if contador==2 : return True
    return False

print(es_primo(5))
````

## Ejercicio: Distancia entre Dos Puntos

Implementa una función que calcule la distancia euclideana entre dos puntos en un espacio n-dimensional.

**Especificaciones:**
- Define una función llamada `distancia(punto1, punto2)` que reciba dos puntos como listas/tuplas
- La función debe validar que ambos puntos tengan la misma dimensión
- Calcula la distancia usando la fórmula: d = √((x₂-x₁)² + (y₂-y₁)² + ... + (nₙ-n₁)²)
- Retorna la distancia como un número real
- Haz dos versiones:
    - La primera con **Python puro** sin librerías externas como NumPy
    - En la segunda utiliza numpy para simplificar la función

**Ejemplo:** `distancia([0, 0], [3, 4])` deberá retornar `5.0`

````python
# hacer
import numpy as np
def distancia(punto1, punto2):
    if len(punto1) != len(punto2): return "Los puntos deben tener la misma dimensión"
    suma = 0
    for i in range(len(punto1)):
        suma += (punto2[i] - punto1[i])**2
    return (suma**0.5)

def distanciaV2(punto1, punto2):
    p1 = np.array(punto1)
    p2 = np.array(punto2)
    if p1.shape != p2.shape:
        raise ValueError("Los puntos deben tener la misma dimensión")
    
    return np.linalg.norm(p2 - p1)

    
print(distancia([0, 0],[3, 4]))
print("- - - - ")
print(distanciaV2([0, 0],[3, 4]))
````
