```python
import numpy as np

def trapecioCompuesto(y,x):
    n = len(y) - 1
    h = (x[-1] - x[0])/n
    I = y[0]+y[-1] + 2*y[1:-1].sum()
    I *= h/2
    return I

def Punto4(a,b,f):
    cant = 0
    n = 2
    tolerancia_trap = 10**-6
    error_trapecio = tolerancia_trap + 1
    x = np.linspace(a,b,n+1)
    y = f(x)
    valor_ant = trapecioCompuesto(y,x)
    while(error_trapecio > tolerancia_trap):
        n += 2
        x = np.linspace(a,b,n+1)
        y = f(x)
        valor = trapecioCompuesto(y,x)
        error_trapecio = abs(valor-valor_ant)
        valor_ant = valor
    
    print("El numero de evaluaciones de la funcion es: ", n+1)

    
def f3(x):
    return np.sin(10*x)
def f4(x):
    return np.e**x**2
def f5(x):
    return np.log(x)

print("--- Función 3 ---")
a1,b1 = 0,2
Punto4(a1,b1,f3)
print("--- Función 4 ---")
a2,b2 = 0,1
Punto4(a2,b2,f4)
print("--- Función 5 ---")
a3,b3 = 1,10
Punto4(a3,b3,f5)
```
