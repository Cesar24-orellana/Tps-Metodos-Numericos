# 3) Regla de Simpson compuesta

```python
import numpy as np

def simpsons_compuesto(x,y):
    n = len(y) -1   # Cantidad de intervalos
    h = (x[-1] - x[0]) / n
    I = y[0]+y[-1]
    for i in range(1,n):
        if i%2 == 0:
            i += 2*y[i]
        else:
            i += 4*y[i]
    I *= h / 3
    return I

def Punto3(f, a, b):    # f: Funcion ; a, b: intervalos cerrado
    error = 10**(-6)
    n = 2
    x = np.linspace(a, b, n+1)
    y = f(x)
    valor_anterior = simpsons_compuesto(x,y)
    tolerancia = 1 - error
    while(tolerancia > error):
        n += 2
        x = np.linspace(a,b, n+1)
        y = f(x)
        valor = simpsons_compuesto(x,y)
        tolerancia = abs(valor - valor_anterior)
        valor_anterior = valor
    print(f"La integral aproximada es: {valor:.8f}")
    print(f"La cantidad de puntos requeridos es: {n}")
```

# 4) Regla del Trapecio Compuesta

```python
import numpy as np

def trapecio_compuesto(x,y):
    n=len(y)-1
    h = (x[-1] - x[0]) / n
    I = y[0] + y[-1] + 2*y[1:-1].sum()
    I *= h / 2
    return I

def Punto4(f,a,b, tol=1e-6):  # f: funcion; a, b: Intervalo de prueba
    n = 2
    x = np.linspace(a,b,n+1)
    y = f(x)
    error = tol*2
    valor_anterior = trapecio_compuesto(x,y)
    while(error > tol):
        n += 2
        x = np.linspace(a,b,n+1)
        y = f(x)
        valor = trapecio_compuesto(x,y)
        error = abs(valor - valor_anterior)
        valor_anterior = valor
    print("El numero de evaluaciones de la funcion es: ", n+1)
```

# 5) Cuadratura de Gauss

```python
import numpy as np

def f1(x):
    return (np.exp(x) * np.sin(x)) / (1 + x**2)

def cuadratura_Gauss(f, a, b, n):
    if n == 2:
        t = np.array([-1/np.sqrt(3), 1/np.sqrt(3)])
        alfa = np.array([1,1])
    elif n == 3:
        t = np.array([-np.sqrt(3/5), 0, np.array(3/5)])
        alfa = np.array([5/9, 8/9, 5/9])
    else:
        raise ValueError("Solo configurado para n=2 o n=3")
    
    dx = (b-a)/2
    x = dx*t + (a+b)/2

    integral = dx* np.sum(alfa * f(x))
    return integral
```

```python
import numpy as np

a,b = 0, 2

I_g2  = cuadratura_Gauss(f1, a, b, 2)
I_g3 = cuadratura_Gauss(f1, a, b, 3)
print("---RESULTADOS CUADRATURA DE GAUSS ---") 
print(f"Limite de integración: {a}, {b}")
print(f"Aproximación para n=2: {I_g2:.4f}")
print(f"Aproximación para n=3: {I_g3:.4f}")
```
