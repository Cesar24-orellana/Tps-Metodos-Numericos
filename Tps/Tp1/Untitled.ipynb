# ii) √(𝑥 + (1/𝑥)) − √(𝑥 − (1/𝑥))


```python
import numpy as np
import matplotlib.pyplot as plt

def f_original(x):
    return np.sqrt(x+(1/x)) - np.sqrt(x-(1/x))

def f_nueva(x):
    return (2*x/(np.sqrt(x+1/x) + np.sqrt(x-1/x)))

eps = 1e-3
x = np.linspace( 1,5, 1000, dtype=np.float32)

plt.figure(figsize=(12,6))
plt.plot(x, f_original(x), label="original")
plt.plot(x, f_nueva(x), label="nueva")
plt.plot(x, f_original(x) - f_nueva(x), label="error")
plt.legend()
plt.show()
```

# iii) x - sen(x)

```python
import numpy as np
import matplotlib.pyplot as plt

def f_Original(x):
    return x - np.sin(x)

def f_Nueva(x):
    return x**3/6 - x**5/120

eps = 1e-3
x = np.linspace(0 - eps, 0 + eps, 100, dtype= np.float32)

plt.plot(x, f_Original(x), label="Originañ")
plt.plot(x,f_Nueva(x), label="Nueva")
plt.legend()
```

# iv) log x - 1

```python
import numpy as np
import matplotlib.pyplot as plt

def F_original(x):
    return np.log(x) - 1

def F_nueva(x):
    return np.log(x/10)

eps = 1e-3
x = np.linspace(3 - eps, 0 + eps, 1000, dtype=np.float32)

plt.plot(x, F_original(x), label= "Original")
plt.plot(x, F_nueva(x), label="Nueva")
plt.legend()
```

# 6) Serie Geometrica

```python
import numpy as np

def serie_geo(n):
    result = 0
    for i in range(1, n + 1):
        result = result + (1/i**4)
    return result

def serie_ascendente(n):
    result = np.float32(0.0)
    for i in range(1, n + 1):
        result = np.float32(result + (1 / i**4))
    return result

def serie_desendente(n):
    result = np.float32(0.0)
    for i in range(n, 0, -1):
        result = np.float32(result + (1 / i**4))
    return result

n = 10000
valor_real = np.pi**4/90
goe = serie_geo(n)
asc = serie_ascendente(n)
des = serie_desendente(n)

er_asc = np.abs(valor_real - asc) / valor_real
er_des = np.abs(valor_real - des) / valor_real

print(f"Error relativa calculado de forma ascendente: {er_asc}")
print(f"Error relativa calculado de forma descendente: {er_des}")
print("Esto se debe a que en el caso descendente sumo terminos de magnitud similar, minimizando la perdida de digitos significativos.")
```

```python
import numpy as np

import math as mt

def aprox_uno(x, n):
    result = 0
    for i in range(n):
        result += (-x)**i / mt.factorial(i)
    return result

def aprox_dos(x, n):
    result = 0
    for i in range(n):
        result += x**i / mt.factorial(i)
    result = 1/ result
    return result

x = 8.3
valor_real = 2.485168e-4

print(f"Valor real: {valor_real}")
print(f"Serie 1 (4 terminos): {aprox_uno(x, 4)}")
print(f"Serie 2 (4 terminos): {aprox_dos(x, 4)}")

print("- - - - - - -")
print(f"Serie 1 (20 terminos): {aprox_uno(x, 20)}")
print(f"Serie 2 (20 terminos): {aprox_dos(x, 20)}")

print("- - - - - - -")
print(f"Serie 1 (2 terminos): {aprox_uno(x, 2)}")
print(f"Serie 2 (2 terminos): {aprox_dos(x, 2)}")


```
